```python
# Data Visualization on Netflix Movies & TV Shows Dataset

This project uses visual charts to understand content distribution,
countries, genres, and trends on Netflix.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("netflix_titles.csv.csv")
df.head(10)
```

```python
df.dropna(subset=['country', 'rating', 'listed_in'], inplace=True)
```

```python
df.info()
```

```python
sns.countplot(x='type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
Netflix hosts significantly more Movies than TV Shows.
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='rating', data=df)
plt.xticks(rotation=45)
plt.title("Content Distribution by Rating")
plt.show()
```

```python
Most Netflix content is rated for mature audiences.
```

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar', title="Top 10 Countries with Most Content")
plt.show()
```

```python
USA and India contribute the most content on Netflix.
```

```python
year_data = df['release_year'].value_counts().sort_index()
year_data.plot(kind='line', title="Content Release Trend Over Years")
plt.show()
```

```python
Content production increased rapidly after 2015.
```

```python
top_genres = df['listed_in'].value_counts().head(10)
top_genres.plot(kind='bar', title="Top 10 Genres on Netflix")
plt.show()
```

```python
Drama and International Movies dominate Netflix genres.
```

```python
# Clean the column first
df['date_added'] = df['date_added'].astype(str).str.strip()

# Convert to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year
df['year_added'] = df['date_added'].dt.year

# Plot
added_year = df['year_added'].value_counts().sort_index()
added_year.plot(kind='bar', title="Content Added to Netflix Each Year")
plt.xlabel("Year")
plt.ylabel("Number of Shows Added")
plt.show()
```

```python
Netflix added maximum content between 2018 and 2020.
```

```python
## Conclusion

Visualizations reveal how Netflix content varies by type, country,
genre, rating, and time. Charts clearly demonstrate trends and
content distribution for storytelling with data.
```
